```python
import pandas as pd
import numpy as np

print("Libraries imported successfully")
```

```python
# Load the data with correct encoding
df = pd.read_csv('../data/raw/Sample - Superstore.csv', encoding='ISO-8859-1')

print("Dataset Shape:", df.shape)
print("Columns:", df.columns.tolist())
display(df.head(3)) 
```

```python
print("Missing Values:\n", df.isnull().sum())
print("\nDuplicate Rows:", df.duplicated().sum())
print("\nData Types:\n", df.dtypes)
```

```python
# Convert date columns
df['Order Date'] = pd.to_datetime(df['Order Date'], errors='coerce')
df['Ship Date'] = pd.to_datetime(df['Ship Date'], errors='coerce')

# Remove duplicates
df = df.drop_duplicates()

# Handle missing values
df['Postal Code'] = df['Postal Code'].fillna('Unknown')

# Drop rows with critical missing values
df = df.dropna(subset=['Sales', 'Profit', 'Order Date'])

print("Shape after basic cleaning:", df.shape)
```

```python
df['Year'] = df['Order Date'].dt.year
df['Month'] = df['Order Date'].dt.month_name()
df['Quarter'] = 'Q' + df['Order Date'].dt.quarter.astype(str)
df['Month-Year'] = df['Order Date'].dt.to_period('M').astype(str)
df['Day of Week'] = df['Order Date'].dt.day_name()

df['Profit Margin %'] = round((df['Profit'] / df['Sales']) * 100, 2)
df['Shipping Days'] = (df['Ship Date'] - df['Order Date']).dt.days
df['Segment-Category'] = df['Segment'] + " - " + df['Category']

# Save cleaned file
df.to_csv('../data/cleaned/Superstore_Cleaned.csv', index=False) 
print("Cleaned dataset saved successfully in data/cleaned/")
```
